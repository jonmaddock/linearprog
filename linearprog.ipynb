# Linear Programming in Python
## A Simple example
Find the independent **decision variables** $x$ and $y$ which maximise the objective function $z$:
\begin{equation*}
z = x + 2y
\end{equation*}

Subject to the **inequality constraints**:
\begin{equation*}
2x + y <= 20
\end{equation*}

\begin{equation*}
-4x + 5y <= 10
\end{equation*}

\begin{equation*}
-x + 2y >= -2
\end{equation*}

\begin{equation*}
x >= 0
\end{equation*}

\begin{equation*}
y >= 0
\end{equation*}

Rearranging the constraints for y:
\begin{equation*}
y <= 20 - 2x
\end{equation*}

\begin{equation*}
y <= \frac{10 + 4x}{5}
\end{equation*}

\begin{equation*}
y >= \frac{-2 + x}{2}
\end{equation*}

This can be plotted:

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10)

constr_a = 20 - 2 * x
constr_b = (10 + 4 * x) / 5
constr_c = (-2 + x) / 2

fig, ax = plt.subplots()
ax.set_title("Simple linear programming example")
ax.set_ylabel("y")
ax.set_xlabel("x")
ax.set_xlim(xmin=x[0], xmax=x[-1])
ax.set_ylim(0, 10)
ax.plot(x, constr_a, "r")
ax.fill_between(x, constr_a, max(constr_a), facecolor="r", alpha=0.3)
ax.plot(x, constr_b, "b")
ax.fill_between(x, constr_b, max(constr_b), facecolor="b", alpha=0.3)
ax.plot(x, constr_c, "y")
ax.fill_between(x, constr_c, 0, facecolor="y", alpha=0.3)
```

The remaining white region is called the **feasible region**; each point in this area satisfies all constraints and is a potential solution to the problem (**feasible solution**). The point which maximises $z$ is called the **optimal solution**. As $z$ is linear it can be visualised as a plane in 3D space. The optimal solution can therefore lie on a vertex, but multiple optimal solutions may exist on an edge or even the entire plane.

Adding an additional **equality constraint**:
$$
-x + 5y = 15
$$

```python
constr_d = 3 + (x / 5)
ax.plot(x, constr_d, "g")
# Show plot again
fig
```

As the green line is an equality constraint, the solution must now lie on it, within the bounded region. 

However, if the values of $x$ must be integers, then the problem becomes a mixed-integer linear programming problem:

```python
x_int = np.arange(0, 11, 1)
constr_d_int = 3 + (x_int / 5)
ax.plot(x_int, constr_d_int, "go")
fig
```

Now only the green points are feasible solutions.

The examples so far have been feasible linear programming problems, as they have bounded feasible regions and finite solutions.
